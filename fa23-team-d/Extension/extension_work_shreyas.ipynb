```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
census = pd.read_csv('data/redistricting_data_tract20_nbhd_hhpopsize_ab-1.csv')
capital = pd.read_csv('data/fy24-capital-budget-plan-recommended.csv')
```

```python
census.head()
```

```python
capital.head()
```

```python
census.columns
```

```python
census.dropna(axis=1)
```

```python
census['P0050008'] = pd.to_numeric(census['P0050008'], errors='coerce')
student_housing = census.groupby('tract20_nbhd')['P0050008'].sum().reset_index()
student_housing = student_housing.sort_values(by='P0050008', ascending=False)
student_housing_fil = student_housing[student_housing['P0050008']!=0]
student_housing_fil.columns = ['Neighborhood', 'Student Housing']
student_housing_fil.head(20)
```

```python
student_housing_fil.set_index('Neighborhood', inplace=True)
student_housing_fil.plot(kind='bar', figsize=(15,10))
plt.title('Student housings across neighborhoods')
plt.ylabel('Student Housing')
plt.xlabel('Neighborhood')
plt.show()
```

```python
capital['Total_Project_Budget'] = pd.to_numeric(capital['Total_Project_Budget'], errors='coerce')
neighborhood  = capital.groupby('Neighborhood')['Total_Project_Budget'].sum().reset_index()
neighborhood = neighborhood.sort_values(by='Total_Project_Budget', ascending=False)
neighborhood_fil = neighborhood[neighborhood['Total_Project_Budget']!=0]
```

```python
neighborhood_fil = neighborhood_fil.drop(neighborhood.index[0])
```

```python
neighborhood_fil.head()
```

```python
people = census[['tract20_nbhd', 'P0020001']]
people.columns = ['Neighborhood', 'Population']
```

```python
# people = people.drop(0, axis=0)
people.columns
```

```python
neighborhood_fil['Neighborhood'] = neighborhood_fil['Neighborhood'].replace('Fenway-Kenmore', 'Fenway')
```

```python
neighborhood_fil.head(20)
```

```python
per_capita = pd.merge(neighborhood_fil, people, on='Neighborhood', how='inner')
```

```python
per_capita['Population'] = pd.to_numeric(per_capita['Population'], errors='coerce')
```

```python
per_capita['Per Capita'] = per_capita['Total_Project_Budget']/per_capita['Population']
per_capita = per_capita.sort_values(by='Per Capita', ascending=False)
```

```python
per_capita.head(20)
```

```python
fig, axs = plt.subplots(figsize=(15,10))
plt.bar(per_capita['Neighborhood'], per_capita['Per Capita'], color='skyblue')
plt.xlabel('Neighborhood')
plt.ylabel('Per Capita')
plt.title('Per capita for each Neighborhood')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
