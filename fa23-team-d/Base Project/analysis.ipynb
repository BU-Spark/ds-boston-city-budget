FY22 Adopted Revenue Budget

```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
from collections import defaultdict

# initialize a list to store fy21 actual data
fy21_actual_values = []
department_sums = defaultdict(int)
line_number = 1

# Read the CSV file into a DataFrame
with open('data/fy24-adopted-revenue-budget.csv', 'r', newline='') as f:
    csv_reader = csv.reader(f)
    header = next(csv_reader)  # Read the header and skip it

    for column in csv_reader:
        fy21_actual = int(column[4].replace(',', ''))
        department = column[3].replace("Department", "").strip() 
        
        if department not in department_sums:
            department_sums[department] = fy21_actual
        else:
            department_sums[department] += fy21_actual

# Sort the departments by total 'FY21 Actual' values in descending order
sorted_departments = sorted(department_sums.items(), key=lambda x: x[1], reverse=True)

# Get the top 5 departments
top_5_departments = sorted_departments[:5]
departments, total_sums = zip(*top_5_departments)

# Create a pie chart
plt.figure(figsize=(9, 8))
plt.pie(total_sums, labels=departments, autopct='%1.1f%%', startangle=40)
plt.title('FY21 Adopted Revenue Budget')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
