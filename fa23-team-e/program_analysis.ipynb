#### Change the columns

```python
operating_budget['FY21 Actual Expense'] = operating_budget['FY21 Actual Expense'].str.replace(',', '')
operating_budget['FY21 Actual Expense'] = pd.to_numeric(operating_budget['FY21 Actual Expense'], errors='coerce')

operating_budget['FY22 Actual Expense'] = operating_budget['FY22 Actual Expense'].str.replace(',', '')
operating_budget['FY22 Actual Expense'] = pd.to_numeric(operating_budget['FY22 Actual Expense'], errors='coerce')

operating_budget['FY23 Appropriation'] = operating_budget['FY23 Appropriation'].str.replace(',', '')
operating_budget['FY23 Appropriation'] = pd.to_numeric(operating_budget['FY23 Appropriation'], errors='coerce')

operating_budget['FY24 Adopted'] = operating_budget['FY24 Adopted'].str.replace(',', '')
operating_budget['FY24 Adopted'] = pd.to_numeric(operating_budget['FY24 Adopted'], errors='coerce')

operating_budget.info()
```

```python
capital_budget['Total_Project_Budget'] = capital_budget['Total_Project_Budget'].str.replace(',', '')
capital_budget['Total_Project_Budget'] = pd.to_numeric(capital_budget['Total_Project_Budget'], errors='coerce')
capital_budget['Total_Project_Budget'] = capital_budget['Total_Project_Budget'].astype('int64')

capital_budget.info()
```

#### Visualizations by Program

```python
operating_budget['FY21-23 Average'] = operating_budget[['FY21 Actual Expense', 'FY22 Actual Expense', 'FY23 Appropriation']].mean(axis=1)
program_frequency = operating_budget.groupby('Program')['FY21-23 Average'].count().sort_values(ascending=False)
print("There are", operating_budget.loc[:, "Program"].nunique(), "unique programs in the operating budget.")
program_frequency.plot(kind='bar')
plt.title('Frequency of Programs')
plt.xticks([])
plt.ylabel('Frequency')
plt.xlabel('Programs')
plt.show()
```

```python
spending_by_program = operating_budget.groupby('Program')['FY21-23 Average'].sum().sort_values(ascending=False)
fig, ax = plt.subplots()
ax.pie(spending_by_program, labels=spending_by_program.index, autopct='%1.1f%%', shadow=False, startangle=0)
fig.set_facecolor('white')
plt.title('Average Spending by Program FY21-23')
plt.show()
```

```python
spending_by_program = operating_budget.groupby('Program')['FY24 Adopted'].sum().sort_values(ascending=False)
fig, ax = plt.subplots()
ax.pie(spending_by_program, labels=spending_by_program.index, autopct='%1.1f%%', shadow=False, startangle=0)
fig.set_facecolor('white')
plt.title('Average Spending by Program FY21-23')
plt.show()
```

```python
spending_by_program = operating_budget.groupby('Program')['FY21-23 Average'].sum().sort_values(ascending=False)
spending_by_program.plot(kind='bar')
plt.title('Average Spending by Program between FY21-23')
plt.ylabel('Amount')
plt.xlabel('Programs')
plt.xticks([])
plt.show()
```

```python
operating_budget['Projected vs Real'] = (operating_budget['FY24 Adopted'] - operating_budget['FY21-23 Average'])
spending_by_program = operating_budget.groupby('Program')['Percent Projected vs Real'].sum().sort_values(ascending=False)
spending_by_program.plot(kind='bar')
plt.title('Percent Increase in Projected Spending vs Actual Spending')
# plt.xticks([])
plt.show()
```

```python
budget_by_time = operating_budget.iloc[:, 4:].sum().sort_values(ascending=True)
budget_by_time.plot(kind='bar')
plt.title('Budget over Time')
plt.ylabel('Total Budget')
plt.show()
```
