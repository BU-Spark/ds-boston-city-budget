There are Missing values that are marked as strings instead of NA. We need make the integer columns correct.

#### Change the columns

```python
operating_budget['FY21 Actual Expense'] = operating_budget['FY21 Actual Expense'].str.replace(',', '')
operating_budget['FY21 Actual Expense'] = pd.to_numeric(operating_budget['FY21 Actual Expense'], errors='coerce')

operating_budget['FY22 Actual Expense'] = operating_budget['FY22 Actual Expense'].str.replace(',', '')
operating_budget['FY22 Actual Expense'] = pd.to_numeric(operating_budget['FY22 Actual Expense'], errors='coerce')

operating_budget['FY23 Appropriation'] = operating_budget['FY23 Appropriation'].str.replace(',', '')
operating_budget['FY23 Appropriation'] = pd.to_numeric(operating_budget['FY23 Appropriation'], errors='coerce')

operating_budget['FY24 Adopted'] = operating_budget['FY24 Adopted'].str.replace(',', '')
operating_budget['FY24 Adopted'] = pd.to_numeric(operating_budget['FY24 Adopted'], errors='coerce')

operating_budget.info()
```

```python
capital_budget['Total_Project_Budget'] = capital_budget['Total_Project_Budget'].str.replace(',', '')
capital_budget['Total_Project_Budget'] = pd.to_numeric(capital_budget['Total_Project_Budget'], errors='coerce')
capital_budget['Total_Project_Budget'] = capital_budget['Total_Project_Budget'].astype('int64')

capital_budget.info()
```

```python

# Total Spending by Department for FY24:
spending_by_department = operating_budget.groupby('Dept')['FY24 Adopted'].sum().sort_values(ascending=False)

spending_by_department.plot(kind='bar')
plt.title('Total Spending by Department for FY24')
plt.ylabel('Amount')
plt.show()
```

```python
spending_by_category = operating_budget.groupby('Expense Category')['FY24 Adopted'].sum().sort_values(ascending=True)

total = sum(spending_by_category)
for x in range(len(spending_by_category)):
    spending_by_category[x] = spending_by_category[x] / total

spending_by_category.plot(kind='pie')
plt.title('Total Spending by Category for FY24')
plt.ylabel('Amount')
plt.show()
```

```python
import numpy as np

y21 = operating_budget.groupby('Expense Category')['FY21 Actual Expense'].sum().sort_values(ascending=True)
y22 = operating_budget.groupby('Expense Category')['FY22 Actual Expense'].sum().sort_values(ascending=True)
y23 = operating_budget.groupby('Expense Category')['FY23 Appropriation'].sum().sort_values(ascending=True)
y24 = operating_budget.groupby('Expense Category')['FY24 Adopted'].sum().sort_values(ascending=True)

x = y22.keys()
y1 = []
y2 = []
y3 = []
y4 = []
for k in y22.keys():
    y1.append(y21[k])
    y2.append(y22[k])
    y3.append(y23[k])
    y4.append(y24[k])


ind = np.arange(7)
width = 0.1

plt.figure(figsize=(20,10))


plt.bar(ind - width * 3, y1 , width * 2, label='FY21')
plt.bar(ind - width * 1, y2 , width * 2, label='FY22')
plt.bar(ind + width * 1, y3, width * 2, label='FY23')
plt.bar(ind + width * 3, y4, width * 2, label='FY24')

plt.xticks(ind, x) 
plt.xlabel("Category", fontsize = 20)
plt.title('Total Spending by Category for FY21 vs FY22 vs FY23 vs FY24', fontsize = 20)
plt.ylabel('Amount', fontsize = 20)
plt.show()
```

```python
y23 = operating_budget.groupby('Expense Category')['FY23 Appropriation'].sum().sort_values(ascending=True)
y24 = operating_budget.groupby('Expense Category')['FY24 Adopted'].sum().sort_values(ascending=True)

x = y23.keys()
y1 = []
y2 = []
for k in y23.keys():
    y1.append(y23[k])
    y2.append(y24[k])

ind = np.arange(7)
width = 0.2

plt.figure(figsize=(20,10))

plt.bar(ind - width, y1 , width * 2, label='FY23')
plt.bar(ind + width, y2, width * 2, label='FY24')

plt.xticks(ind, x) 
plt.xlabel("Category", fontsize = 20)
plt.title('Total Spending by Category for FY23 vs FY24', fontsize = 20)
plt.ylabel('Amount', fontsize = 20)
plt.show()
```
