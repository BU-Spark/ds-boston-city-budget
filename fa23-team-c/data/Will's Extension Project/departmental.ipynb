```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import os

data_directory = 'data/'

file_names = [f"checkbook_explorerfy{fy}.csv" for fy in range(12, 24)]
```

```python

all_data_corrected = pd.DataFrame()

for file_name in file_names:
    file_path = os.path.join(data_directory, file_name)
    try:
        fiscal_year = int('20' + file_name[-6:-4])
        print(f"Loading file {file_name} with fiscal year {fiscal_year}")

        if fiscal_year >= 2021:
            df = pd.read_csv(file_path, encoding='ISO-8859-1')
        else:
            df = pd.read_csv(file_path)

        df['fiscal_year'] = fiscal_year

        all_data_corrected = pd.concat([all_data_corrected, df], ignore_index=True)
    except Exception as e:
        print(f"Error loading file {file_name}: {e}")

all_data_corrected.head()
```

```python
import pandas as pd
import os

all_data_corrected = pd.DataFrame()

for file_name in file_names:
    file_path = os.path.join(data_directory, file_name)
    try:
        fiscal_year = int('20' + file_name[-6:-4])
        encoding = 'ISO-8859-1' if fiscal_year >= 2021 else None
        df = pd.read_csv(file_path, encoding=encoding)
        
        df.columns = df.columns.str.lower().str.replace(' ', '_')

        df['fiscal_year'] = fiscal_year

        all_data_corrected = pd.concat([all_data_corrected, df], ignore_index=True)

    except Exception as e:
        print(f"Error loading file {file_name}: {e}")

all_data_corrected.head()
```

# Total Trend

```python

all_data_corrected['Monetary_Amount'] = pd.to_numeric(all_data_corrected['Monetary_Amount'], errors='coerce')

yearly_spending_corrected = all_data_corrected.groupby('Fiscal_Year')['Monetary_Amount'].sum()

plt.figure(figsize=(10, 6))
sns.lineplot(x=yearly_spending_corrected.index, y=yearly_spending_corrected.values)
plt.title('Trend of Total Spending by Fiscal Year')
plt.xlabel('Fiscal Year')
plt.ylabel('Total Spending (in billions)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Trend by department

```python

department_yearly_spending = all_data_corrected.groupby(['Dept_Name', 'Fiscal_Year'])['Monetary_Amount'].sum().reset_index()

department_yearly_spending_pivot = department_yearly_spending.pivot(index='Fiscal_Year', columns='Dept_Name', values='Monetary_Amount')

department_yearly_spending_pivot = department_yearly_spending_pivot.fillna(0)

top_departments = department_yearly_spending.groupby('Dept_Name')['Monetary_Amount'].sum().nlargest(10).index
top_department_spending_pivot = department_yearly_spending_pivot[top_departments]

# Plotting the trend of spending over the years for top departments
plt.figure(figsize=(15, 10))
sns.lineplot(data=top_department_spending_pivot)
plt.title('Trend of Spending by Department Over Fiscal Years')
plt.xlabel('Fiscal Year')
plt.ylabel('Total Spending (in billions)')
plt.xticks(rotation=45)
plt.legend(title='Department', labels=top_departments, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Normalized Trend

```python

total_city_spending_per_year = all_data_corrected.groupby('Fiscal_Year')['Monetary_Amount'].sum()

normalized_department_spending = department_yearly_spending.pivot(index='Fiscal_Year', columns='Dept_Name', values='Monetary_Amount')
normalized_department_spending = normalized_department_spending.divide(total_city_spending_per_year, axis=0)

normalized_top_department_spending = normalized_department_spending[top_departments]

plt.figure(figsize=(15, 10))
sns.lineplot(data=normalized_top_department_spending)
plt.title('Normalized Spending Trend of Top Departments Over Fiscal Years')
plt.xlabel('Fiscal Year')
plt.ylabel('Proportion of Total City Spending')
plt.xticks(rotation=45)
plt.legend(title='Department', labels=top_departments, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Monthly Spending Pattern

```python

monthly_spending = all_data_corrected.groupby(['Month', 'Fiscal_Year'])['Monetary_Amount'].sum().reset_index()


month_map = {'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June': 6, 
             'July': 7, 'August': 8, 'September': 9, 'October': 10, 'November': 11, 'December': 12}
monthly_spending['Month_Num'] = monthly_spending['Month'].map(month_map)


monthly_spending_sorted = monthly_spending.sort_values(by=['Fiscal_Year', 'Month_Num'])

monthly_spending_pivot = monthly_spending_sorted.pivot(index='Fiscal_Year', columns='Month', values='Monetary_Amount')

plt.figure(figsize=(15, 10))
sns.heatmap(monthly_spending_pivot, cmap="YlGnBu", annot=False, cbar_kws={'label': 'Total Spending (in billions)'})
plt.title('Monthly Spending Patterns Over Fiscal Years')
plt.xlabel('Month')
plt.ylabel('Fiscal Year')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
