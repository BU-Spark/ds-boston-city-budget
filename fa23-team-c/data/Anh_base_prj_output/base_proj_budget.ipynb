```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read in csv file
def get_data(filepath):
    budget = pd.read_csv(filepath)
    return budget
budget = get_data('fy24-adopted-operating-budget.csv')
budget['FY21 Actual Expense'] = pd.to_numeric(budget['FY21 Actual Expense'], errors='coerce')
budget['FY22 Actual Expense'] = pd.to_numeric(budget['FY22 Actual Expense'], errors='coerce')
budget['FY23 Appropriation'] = pd.to_numeric(budget['FY23 Appropriation'], errors='coerce')
budget['FY24 Adopted'] = pd.to_numeric(budget['FY24 Adopted'], errors='coerce')
```

```python
#Department Spending FY21
ax = budget.groupby('Dept')['FY21 Actual Expense'].sum().plot.bar(xlabel ='Dept', ylabel='Spending', title='Department Spending FY21', figsize=(12, 10))
dept = budget.groupby('Dept')['FY21 Actual Expense'].sum()
dept = dept.sort_values(ascending=False)
print(dept)
```
