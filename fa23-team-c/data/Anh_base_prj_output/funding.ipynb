```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df21 = pd.read_csv('checkbook_explorerfy21.csv', encoding='latin1')
df22 = pd.read_csv('checkbook_explorerfy22.csv', encoding='latin1')
df23 = pd.read_csv('checkbook_explorerfy23.csv', encoding='latin1')
df24 = pd.read_csv('checkbook_explorefy24.csv', encoding='latin1')
```

```python
df21.dropna(inplace=True)
account_descr21 = df21['Account_Descr'].unique()
account_descr21 = df21.groupby('Account_Descr')['Monetary_Amount'].sum()
account_descr21 = account_descr21.sort_values(ascending=False).head(10)
account_descr21.plot(kind='bar')
plt.title('Account Description Funding 2021')
plt.xlabel('Account Description')
plt.ylabel('Funding')
plt.show()
```

```python
dept21 = df21.groupby('Dept_Name')['Monetary_Amount'].sum()
dept21
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

# Plot for 2021
top5_2021 = dept21.sort_values(ascending=False).head(5)
axes[0, 0].bar(top5_2021.index, top5_2021.values)
axes[0, 0].set_title('Top 10 Departments by Funding in 2021')
axes[0, 0].set_xlabel('Department')
axes[0, 0].set_ylabel('Funding')

# Plot for 2022
top5_2022 = dept22.sort_values(ascending=False).head(5)
axes[0, 1].bar(top5_2022.index, top5_2022.values)
axes[0, 1].set_title('Top 10 Departments by Funding in 2022')
axes[0, 1].set_xlabel('Department')
axes[0, 1].set_ylabel('Funding')

# Plot for 2023
top5_2023 = dept23.sort_values(ascending=False).head(5)
axes[1, 0].bar(top5_2023.index, top5_2023.values)
axes[1, 0].set_title('Top 10 Departments by Funding in 2023')
axes[1, 0].set_xlabel('Department')
axes[1, 0].set_ylabel('Funding')

# Plot for 2024
top5_2024 = dept24.sort_values(ascending=False).head(5)
axes[1, 1].bar(top5_2024.index, top5_2024.values)
axes[1, 1].set_title('Top 10 Departments by Funding in 2024')
axes[1, 1].set_xlabel('Department')
axes[1, 1].set_ylabel('Funding')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 10))

# Plot for 2021
top5_2021 = dept21.sort_values(ascending=False).head(10)
ax.barh(top5_2021.index, top5_2021.values, label='2021', color='blue')

# Plot for 2022
top5_2022 = dept22.sort_values(ascending=False).head(10)
ax.barh(top5_2022.index, top5_2022.values, label='2022', color='green')

# Plot for 2023
top5_2023 = dept23.sort_values(ascending=False).head(10)
ax.barh(top5_2023.index, top5_2023.values, label='2023', color='red')

# Plot for 2024
top5_2024 = dept24.sort_values(ascending=False).head(10)
ax.barh(top5_2024.index, top5_2024.values, label='2024', color='orange')

ax.set_title('Top 10 Departments by Funding')
ax.set_xlabel('Funding')
ax.set_ylabel('Department')
ax.legend()

plt.show()
```
