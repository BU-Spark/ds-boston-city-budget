```python
import pandas as pd
import numpy as np
df = pd.read_csv("fy24-capital-budget-plan-recommended.csv")
df.head()
```

```python
df_race = pd.read_csv('neighborhoodpop.csv')
df_race.columns = df_race.iloc[1]
df_race = df_race.iloc[2:]
df_race = df_race.iloc[:, :-2]
df_race = df_race.rename(columns={df_race.columns[0]: 'Neighborhood'})
df_race.head()
```

```python
df = df.replace("#Missing", np.nan)
df = df.fillna(0)
df.drop_duplicates

df['Total_Project_Budget'] = pd.to_numeric(df['Total_Project_Budget'], errors='coerce')
```

```python
df_sorted = df.sort_values(by=['Total_Project_Budget'], ascending=False)
df_sorted.head(10)
```

```python
budgets = df.groupby(by = 'Neighborhood', as_index = False)['Total_Project_Budget'].sum()
budgets.sort_values(by=['Total_Project_Budget'], inplace=True)
df = df.drop(df[df['Neighborhood'] == 'Citywide'].index)
df = df.drop(df[df['Neighborhood'] == 'Multiple Neighborhoods'].index)
budgets
```

```python
df = df.groupby(['Neighborhood', 'PM_Department'], as_index=False)['Total_Project_Budget'].sum()
```

```python
from matplotlib import pyplot as plt


plt.figure(figsize=(20,10))
for i, department in enumerate(df['PM_Department'].unique()):
    dept_data = df[df['PM_Department'] == department]
    plt.bar(dept_data['Neighborhood'], dept_data['Total_Project_Budget'], label=department)

plt.xlabel('Neighborhood')
plt.ylabel('Total Project Budget')
plt.title('Total Project Budget by Department and Neighborhood')
plt.legend(title='PM_Department', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=90)  
plt.tight_layout()  
plt.show()
```

```python
neighborhoods = df['Neighborhood'].unique()

for neighborhood in neighborhoods:
    neighborhood_data = df[df['Neighborhood'] == neighborhood]
    percentage = neighborhood_data['Total_Project_Budget'] / neighborhood_data['Total_Project_Budget'].sum() * 100

    fig, ax = plt.subplots()
    wedges, texts, autotexts = ax.pie(percentage, autopct='%1.1f%%')
    ax.set_title(neighborhood)
    ax.legend(wedges, neighborhood_data['PM_Department'], title="Departments", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

    plt.show()
```

```python
df = df[['Neighborhood', 'Total_Project_Budget']]
df = df.groupby(['Neighborhood'], as_index=False)['Total_Project_Budget'].sum()
df = df.merge(df_race, on='Neighborhood')


df
```

```python
new_column_name, column_index_to_rename = 'White %', 4
df.columns.values[column_index_to_rename] = new_column_name

new_column_name, column_index_to_rename = 'Black %', 6
df.columns.values[column_index_to_rename] = new_column_name

new_column_name, column_index_to_rename = 'Hispanic %', 8
df.columns.values[column_index_to_rename] = new_column_name

new_column_name, column_index_to_rename = 'Asian %', 10
df.columns.values[column_index_to_rename] = new_column_name


new_column_name, column_index_to_rename = 'Other %', 12
df.columns.values[column_index_to_rename] = new_column_name

df['Total Population'] = df['Total Population'].str.replace(',', '').astype(float)


df['Budget per Capita'] = df['Total_Project_Budget'] / df['Total Population']

keep = ['Neighborhood', 'Budget per Capita'] + [col for col in df.columns if '%' in col]

df = df[keep]
df
```

```python
df.loc[:, 'White %'] = df['White %'].str.replace('%', '').astype(float)/100
df.loc[:, 'Black %'] = df['Black %'].str.replace('%', '').astype(float)/100
df.loc[:, 'Hispanic %'] = df['Hispanic %'].str.replace('%', '').astype(float)/100
df.loc[:, 'Asian %'] = df['Asian %'].str.replace('%', '').astype(float)/100
df.loc[:, 'Other %'] = df['Other %'].str.replace('%', '').astype(float)/100

df
```

```python
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
# top 3 whitest neighborhoods
top_3 = df.nlargest(3, 'White %')['Neighborhood'].to_frame()
white_top = df[df['Neighborhood'].isin(top_3['Neighborhood'].unique())]

# bottom 3 whitest neighborhoods
bottom_3 = df.nsmallest(3, 'White %')['Neighborhood'].to_frame()
white_bottom = df[df['Neighborhood'].isin(bottom_3['Neighborhood'].unique())]
```
